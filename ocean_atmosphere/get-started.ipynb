```python
import matplotlib.pyplot as plt
import netCDF4
import numpy as np
import sys, glob
```

```python
data_dir = "/net/172.16.118.188/data/ocean_atmosphere/simulations/wmed/processed"
```

```python
oce = sorted(glob.glob(data_dir + "/oce_*.nc"))[-1]
atm = sorted(glob.glob(data_dir + "/atm_*.nc"))[-1]
```

```python
oce = netCDF4.Dataset(oce, "r")
atm = netCDF4.Dataset(atm, "r")
```

```python
plt.pcolormesh(
    oce['oce_xr'],
    oce['oce_yr'],
    oce["oce_sustr"],
    shading='auto',
    cmap='RdBu_r')
plt.colorbar()
```

```python
plt.pcolormesh(
    oce['oce_xr'],
    oce['oce_yr'],
    oce["atm_sustr"],
    shading='auto',
    cmap='RdBu_r')
plt.colorbar()
```
